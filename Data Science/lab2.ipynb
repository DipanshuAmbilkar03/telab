```python
df = df.loc[:,~df.columns.duplicated()]
```

```python
df
```

```python
df.columns = df.columns.str.strip()
```

```python
df.replace(['-', 'Statistics back', ' '], np.nan, inplace=True)
```

```python

# CGPA columns
cgpa_cols = [
    'First Year :  Sem I CGPA',
    'First Year :  Sem II  CGPA',
    'Second Year :  Sem I  CGPA',
    'Second Year :  Sem II  CGPA'
]

for col in cgpa_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df[cgpa_cols] = df[cgpa_cols].fillna(df[cgpa_cols].mean())
```

```python
df[cgpa_cols]
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
# sns.boxplot(data=df,x=cgpa_cols['First Year : Sem I CGPA'])
# plt.show()
```

```python
df['Roll No e.g TE_A_12'] = df['Roll No e.g TE_A_12'].str.replace(' ', '_')

for col in cgpa_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # cap outliers
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
```

```python
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[cgpa_cols] = scaler.fit_transform(df_scaled[cgpa_cols])

print("Cleaned Dataset:")
print(df.head())

print("\nScaled CGPA Dataset:")
print(df_scaled[cgpa_cols].head())
```
