# Data Analytics III - Task 6
Naive Bayes classification on iris.csv with confusion matrix and metrics.

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import LabelEncoder

candidates = [Path('../DS_lab3/Iris.csv'), Path('Iris.csv'), Path('../dataset/iris.csv')]
df = None
for p in candidates:
    if p.exists():
        df = pd.read_csv(p)
        break
if df is None:
    from sklearn.datasets import load_iris
    iris = load_iris(as_frame=True)
    df = iris.frame.copy()

target_col = df.columns[-1]
X = df.iloc[:, :-1].copy()
y = df[target_col].copy()
if y.dtype == 'O':
    y = LabelEncoder().fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
err = 1 - acc
prec = precision_score(y_test, y_pred, average='macro', zero_division=0)
rec = recall_score(y_test, y_pred, average='macro', zero_division=0)

print('Confusion Matrix:')
print(cm)
print('')
rows = []
for i in range(cm.shape[0]):
    tp = cm[i, i]
    fp = cm[:, i].sum() - tp
    fn = cm[i, :].sum() - tp
    tn = cm.sum() - tp - fp - fn
    rows.append([i, int(tp), int(fp), int(tn), int(fn)])

metrics_df = pd.DataFrame(rows, columns=['Class', 'TP', 'FP', 'TN', 'FN'])
print('TP FP TN FN by class:')
print(metrics_df.to_string(index=False))
print('')
print(f'Accuracy: {acc:.4f}')
print(f'Error Rate: {err:.4f}')
print(f'Precision (macro): {prec:.4f}')
print(f'Recall (macro): {rec:.4f}')
```
