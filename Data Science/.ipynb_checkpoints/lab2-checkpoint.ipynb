```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)

df = pd.DataFrame({
    "Student_ID": range(1,21),
    "Name": ["S"+str(i) for i in range(1,21)],
    "Gender": np.random.choice(["Male","Female","female","M"],20),
    "Math_Score": np.random.randint(30,100,20).astype(float),
    "Science_Score": np.random.randint(35,100,20).astype(float),
    "Study_Hours": np.random.randint(1,10,20).astype(float),
    "Attendance": np.random.randint(60,100,20).astype(float)
})

df.loc[3,"Math_Score"] = np.nan
df.loc[5,"Study_Hours"] = np.nan
df.loc[10,"Attendance"] = 200  
df.loc[15,"Science_Score"] = 5 

df["Gender"] = df["Gender"].replace({"female":"Female","M":"Male"})


df.fillna(df.median(numeric_only=True), inplace=True)

num_cols = ["Math_Score","Science_Score","Study_Hours","Attendance"]

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    df[col] = np.where(df[col] > upper, upper,
              np.where(df[col] < lower, lower, df[col]))

df["Study_Hours_Log"] = np.log1p(df["Study_Hours"])

print(df.head())
print("\nSkewness before:", stats.skew(df["Study_Hours"]))
print("Skewness after :", stats.skew(df["Study_Hours_Log"]))

plt.hist(df["Study_Hours_Log"])
plt.title("Transformed Study Hours Distribution")
plt.show()
```
