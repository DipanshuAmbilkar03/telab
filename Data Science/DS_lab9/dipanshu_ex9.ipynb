# Experiment 9: Data Visualization II

Use the inbuilt Titanic dataset and plot a box plot for age distribution by gender with survival information.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid', context='notebook')

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Keep required columns and drop missing ages
df = titanic[['sex', 'age', 'survived']].dropna()
df['survived'] = df['survived'].map({0: 'No', 1: 'Yes'})

print('Shape after cleaning:', df.shape)
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='sex', y='age', hue='survived', palette='Set2')
plt.title('Age Distribution by Sex and Survival (Titanic)')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived')
plt.tight_layout()
plt.show()
```

```python
summary = df.groupby(['sex', 'survived'])['age'].describe().round(2)
summary
```

## Observations / Inference

1. The plot compares age distributions across male and female passengers, split by survival.
2. Survivors tend to have slightly lower median ages in multiple groups, indicating better survival among younger passengers.
3. Male non-survivors generally show a wider spread (larger IQR), which indicates higher age variability.
4. Outliers appear in several groups, showing presence of unusually high or low ages in each category.
5. The visualization supports the conclusion that both sex and age influenced survival outcomes in Titanic.
