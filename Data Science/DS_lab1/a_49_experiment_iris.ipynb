# A-49 Experiment: Data Wrangling I (Iris Dataset)

This notebook completes all required Data Wrangling-I operations on the Iris dataset.

## 1) Import Required Python Libraries

We import libraries for data loading, preprocessing, normalization, and encoding.

```python
import io
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

## 2) Locate Open Source Data From the Web

**Dataset:** Iris (Fisher's Iris Data)  \n
**Source website:** UCI Machine Learning Repository  \n
**URL:** https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

### Dataset Description
- The dataset contains measurements of iris flowers from 3 species.
- Each record has 4 numeric features and 1 categorical target label:
  - sepal_length
  - sepal_width
  - petal_length
  - petal_width
  - species
- Total non-empty records: 150

### How data is imported/read from web
- We pass the direct data URL to `pandas.read_csv(...)`.
- `read_csv` downloads and parses the file in one step.
- We provide custom column names because this raw file has no header row.

```python
dataset_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
    "species"
]

df = pd.read_csv(dataset_url, names=column_names)

# The source has a trailing blank line; remove fully-empty rows if present.
df = df.dropna(how="all").reset_index(drop=True)

print("Dataset loaded successfully from web URL.")
print("Shape (rows, columns):", df.shape)
df.head()
```

## 3) Load Dataset into Pandas DataFrame

The dataset has been loaded into the DataFrame `df` in the previous cell.

## 4) Data Preprocessing

Tasks covered here:
- Missing value check using `isnull()`
- Initial statistics using `describe()`
- Variable descriptions and data types
- DataFrame dimensions

```python
print("DataFrame dimensions (rows, columns):", df.shape)

print("\nMissing values per column:")
missing_counts = df.isnull().sum()
print(missing_counts)

print("\nTotal missing values in dataset:", int(missing_counts.sum()))

print("\nDescriptive statistics for numeric variables:")
display(df.describe())

print("Data types before any conversion:")
display(df.dtypes.to_frame(name="dtype"))

variable_descriptions = pd.DataFrame({
    "Variable": [
        "sepal_length",
        "sepal_width",
        "petal_length",
        "petal_width",
        "species"
    ],
    "Description": [
        "Sepal length in centimeters",
        "Sepal width in centimeters",
        "Petal length in centimeters",
        "Petal width in centimeters",
        "Flower species label (setosa/versicolor/virginica)"
    ],
    "Type": [
        "Numeric (continuous)",
        "Numeric (continuous)",
        "Numeric (continuous)",
        "Numeric (continuous)",
        "Categorical"
    ]
})

print("Variable descriptions:")
display(variable_descriptions)
```

## 5) Data Formatting and Data Normalization

- Summarize variable categories by dtype
- Convert data types where required
- Normalize numeric variables

```python
print("Current dtypes:")
display(df.dtypes.to_frame(name="dtype"))

# Convert species from object (string) to category for better memory efficiency and semantics.
df["species"] = df["species"].astype("category")

print("Dtypes after conversion:")
display(df.dtypes.to_frame(name="dtype"))

dtype_summary = df.dtypes.astype(str).value_counts().rename_axis("dtype").reset_index(name="count")
print("Summary of variable types:")
display(dtype_summary)

numeric_columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
scaler = MinMaxScaler()
df_normalized = df.copy()
df_normalized[numeric_columns] = scaler.fit_transform(df[numeric_columns])

print("First 5 rows after Min-Max normalization (numeric columns scaled to [0,1]):")
display(df_normalized.head())
```

## 6) Convert Categorical Variables to Quantitative Variables

We convert the categorical `species` column into numeric columns using one-hot encoding.

```python
df_encoded = pd.get_dummies(df_normalized, columns=["species"], dtype=int)

print("Shape before encoding:", df_normalized.shape)
print("Shape after one-hot encoding:", df_encoded.shape)
print("\nEncoded columns:")
print([col for col in df_encoded.columns if col.startswith("species_")])

display(df_encoded.head())
```
