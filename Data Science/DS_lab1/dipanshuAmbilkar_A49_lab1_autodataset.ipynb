```python
df['stroke'] = df['stroke'].fillna(method='ffill')
df['horsepower'] = df['horsepower'].fillna(method='ffill') 
df['peak-rpm'] = df['peak-rpm'].fillna(method='ffill') 
df['horsepower-binned'] = df['horsepower-binned'].fillna(method='ffill') 
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.head(3)
```

```python
df['horsepower-binned'].unique()
```

```python
df['aspiration'] = df['aspiration'].map({"std" : 0,"turbo": 1})
```

```python
df['num-of-doors'] = df['num-of-doors'].map({"two" : 2,"four": 4})
```

```python
df['num-of-cylinders'] = df['num-of-cylinders'].map({'four':4, 'six':6, 'five':5, 'three':3, 'twelve':11, 'two' :2, 'eight' : 8}) 
```

```python
df['drive-wheels'] = df['drive-wheels'].map({"rwd" :0, "fwd": 1,"4wd" : 2})
```

```python
df['engine-location'] = df['engine-location'].map({"front" : 0,"rear": 1})
```

```python
df['horsepower-binned'] = df['horsepower-binned'].map({"Low" : 0,"Medium": 1,'High':2})
```
