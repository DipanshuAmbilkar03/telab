```python
import pandas as pd

df = pd.read_csv("iris.csv")

numeric_cols = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]

overall_mean = df[numeric_cols].mean()
overall_median = df[numeric_cols].median()
overall_min = df[numeric_cols].min()
overall_max = df[numeric_cols].max()
overall_std = df[numeric_cols].std()

print("Overall Mean")
print(pd.DataFrame(overall_mean))
print("\nOverall Median")
print(overall_median)
print("\nOverall Minimum")
print(overall_min)
print("\nOverall Maximum")
print(overall_max)
print("\nOverall Standard Deviation")
print(overall_std)

grouped_stats = df.groupby("Class Label")[numeric_cols].agg(["mean","median","min","max","std"])
print("\nGrouped Statistics by Species")
print(grouped_stats)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("iris.csv")

# Select numeric columns
numeric_cols = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]

# ---- Overall Statistics ----
overall_stats = pd.DataFrame({
    "Mean": df[numeric_cols].mean(),
    "Median": df[numeric_cols].median(),
    "Min": df[numeric_cols].min(),
    "Max": df[numeric_cols].max(),
    "Std Dev": df[numeric_cols].std()
})

print("\n========== OVERALL STATISTICS ==========\n")
print(overall_stats)

# ---- Grouped Statistics ----
grouped_stats = df.groupby("Class Label")[numeric_cols].agg(
    ["mean", "median", "min", "max", "std"]
)

print("\n========== GROUPED STATISTICS (BY CLASS LABEL) ==========\n")
print(grouped_stats)
```

```python
import pandas as pd

df = pd.read_csv("iris.csv")

numeric_cols = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]

species_list = df["Class Label"].unique()

for species in species_list:
    print("\nStatistics for:", species)
    subset = df[df["Class Label"] == species][numeric_cols]
    print("Mean")
    print(subset.mean())
    print("\nStandard Deviation")
    print(subset.std())
    print("\nPercentiles")
    print(subset.quantile([0.25,0.5,0.75]))
```

```python
import matplotlib.pyplot as plt
numeric_cols = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]

for col in numeric_cols:
    plt.figure(figsize=(7,5))
    df.boxplot(column=col,by="Class Label")
    plt.title(col+" Boxplot by Species")
    plt.suptitle("")
    plt.xlabel("Species")
    plt.ylabel(col)
    plt.show()
```

```python

for col in numeric_cols:
    plt.figure()
    for label in df["Class Label"].unique():
        subset = df[df["Class Label"]==label]
        plt.hist(subset[col],alpha=0.5,label=label)
    plt.title(col+" Histogram")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
```

```python
plt.figure()
colors = df["Class Label"].astype("category").cat.codes
plt.scatter(df["SepalLengthCm"],df["PetalLengthCm"],c=colors)
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.title("Sepal Length vs Petal Length")
plt.show()
```
