# Data Visualization I - Task 8
Pattern analysis on Titanic dataset and fare distribution histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
print('Shape:', titanic.shape)
display(titanic.head())
display(titanic.describe(include='all'))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.countplot(data=titanic, x='sex', hue='survived', ax=axes[0, 0])
axes[0, 0].set_title('Survival by Sex')
sns.countplot(data=titanic, x='class', hue='survived', ax=axes[0, 1])
axes[0, 1].set_title('Survival by Passenger Class')
sns.boxplot(data=titanic, x='class', y='fare', ax=axes[1, 0])
axes[1, 0].set_title('Fare by Class')
corr = titanic.select_dtypes(include='number').corr()
sns.heatmap(corr, cmap='Blues', ax=axes[1, 1],annot=True)
axes[1, 1].set_title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(titanic['fare'].dropna(), bins=30, kde=True, color='teal')
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
