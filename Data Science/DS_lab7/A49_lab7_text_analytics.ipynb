# Text Analytics - Task 7
Tokenization, POS tagging, stop words removal, stemming, lemmatization, TF and TF-IDF.

```python
import re
import pandas as pd
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS

document = '''
Natural language processing makes it possible for computers to understand text.
Text analytics can extract meaning from customer feedback and social media posts.
Machine learning models use cleaned and transformed text features.
'''

sentences = [s.strip() for s in re.split(r'[.!?]+', document) if s.strip()]
tokens = re.findall(r'[A-Za-z]+', document.lower())

def pos_tag_word(word):
    if word.endswith('ing') or word.endswith('ed'):
        return 'VERB'
    if word.endswith('ly'):
        return 'ADV'
    if word.endswith(('ous', 'ful', 'ive', 'al', 'ic')):
        return 'ADJ'
    return 'NOUN'

def stem_word(word):
    suffixes = ['ingly', 'edly', 'ing', 'ed', 'ly', 'ment', 'ness', 'tion', 's']
    for suf in suffixes:
        if word.endswith(suf) and len(word) > len(suf) + 2:
            return word[:-len(suf)]
    return word

def lemmatize_word(word):
    rules = {'models': 'model', 'features': 'feature', 'posts': 'post', 'computers': 'computer'}
    if word in rules:
        return rules[word]
    if word.endswith('ies') and len(word) > 4:
        return word[:-3] + 'y'
    if word.endswith('s') and len(word) > 3:
        return word[:-1]
    return word

pos_tags = [(w, pos_tag_word(w)) for w in tokens]
stop_words = set(ENGLISH_STOP_WORDS)
filtered_tokens = [t for t in tokens if t not in stop_words]
stemmed_tokens = [stem_word(t) for t in filtered_tokens]
lemmatized_tokens = [lemmatize_word(t) for t in filtered_tokens]

tf_counter = Counter(filtered_tokens)
tf_df = pd.DataFrame({'term': list(tf_counter.keys()), 'frequency': list(tf_counter.values())})
tf_df['tf'] = tf_df['frequency'] / tf_df['frequency'].sum()
tf_df = tf_df.sort_values('frequency', ascending=False).reset_index(drop=True)

clean_docs = []
for s in sentences:
    s_tokens = [w for w in re.findall(r'[A-Za-z]+', s.lower()) if w not in stop_words]
    clean_docs.append(' '.join(s_tokens))

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(clean_docs)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print('Tokens:')
print(tokens)
print('')
print('POS Tags:')
print(pos_tags)
print('')
print('After Stop Words Removal:')
print(filtered_tokens)
print('')
print('Stemmed Tokens:')
print(stemmed_tokens)
print('')
print('Lemmatized Tokens:')
print(lemmatized_tokens)
print('')
print('Term Frequency Table:')
print(tf_df.to_string(index=False))
print('')
print('TF-IDF Matrix:')
print(tfidf_df.to_string(index=False))
```
