```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = [1,2,4,6,9,0,4,34,6]
```

```python
x = np.linspace(-10, 10, 400)

binary_step = np.where(x >= 0, 1, 0)
sigmoid = 1 / (1 + np.exp(-x))
tanh = 2 / (1 + np.exp(-2*x))
relu = np.maximum(0, x)
leaky_relu = np.where(x >= 0, x, 0.01 * x)
```

```python
plt.figure()
plt.plot(x, binary_step, label="Binary Step")
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.plot(x, sigmoid, label="Sigmoid")

plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.plot(x, tanh, label="Tanh")

plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.plot(x, relu, label="ReLU")

plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.plot(x, leaky_relu, label="Leaky ReLU")

plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```
