```python
"""A-3: Perceptron to recognize even and odd digits from ASCII (0-9)."""

import numpy as np


class Perceptron:
    def __init__(self, n_features: int, lr: float = 0.1):
        self.lr = lr
        self.w = np.zeros(n_features)
        self.b = 0.0

    def predict(self, x: np.ndarray) -> int:
        net = np.dot(self.w, x) + self.b
        return 1 if net >= 0 else 0

    def fit(self, x_train: np.ndarray, y_train: np.ndarray, epochs: int = 100):
        for epoch in range(epochs):
            errors = 0
            for x, y in zip(x_train, y_train):
                y_hat = self.predict(x)
                update = self.lr * (y - y_hat)
                self.w += update * x
                self.b += update
                errors += int(update != 0)

            if errors == 0:
                print(f"Converged in {epoch + 1} epochs")
                break


def ascii_to_bits(ch: str) -> np.ndarray:
    return np.array([int(bit) for bit in format(ord(ch), "08b")], dtype=float)


def accuracy(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    return float((y_true == y_pred).mean() * 100)


def main() -> None:
    digits = [str(d) for d in range(10)]
    x = np.array([ascii_to_bits(d) for d in digits], dtype=float)
    # Label: 0 -> even, 1 -> odd
    y = np.array([int(d) % 2 for d in digits], dtype=int)

    model = Perceptron(n_features=8, lr=0.1)
    model.fit(x, y, epochs=100)

    preds = np.array([model.predict(row) for row in x])
    acc = accuracy(y, preds)

    print("\nDigit  ASCII  Predicted")
    print("------------------------")
    for d, p in zip(digits, preds):
        print(f"  {d}     {ord(d)}      {'Odd' if p == 1 else 'Even'}")

    print(f"\nTraining accuracy: {acc:.2f}%")


if __name__ == "__main__":
    main()

```
