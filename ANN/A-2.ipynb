```python
"""A-2: Generate ANDNOT function using a McCulloch-Pitts neural net."""

from itertools import product


def binary_step(net: float) -> int:
    return 1 if net >= 0 else 0


def mcp_and_not(a: int, b: int) -> int:
    # Net = (1 * A) + (-1 * B) + bias.
    w1, w2, bias = 1.0, -1.0, -0.5
    net = w1 * a + w2 * b + bias
    return binary_step(net)


def main() -> None:
    print("McCulloch-Pitts ANDNOT (A AND NOT B)")
    print("A B | Y")
    print("--------")

    correct = 0
    for a, b in product([0, 1], repeat=2):
        y = mcp_and_not(a, b)
        expected = 1 if (a == 1 and b == 0) else 0
        correct += int(y == expected)
        print(f"{a} {b} | {y}")

    accuracy = correct / 4 * 100
    print(f"\nTruth-table accuracy: {accuracy:.2f}%")


if __name__ == "__main__":
    main()

```
