```python
"""A-1: Plot common activation functions used in neural networks."""

import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x: np.ndarray) -> np.ndarray:
    return 1.0 / (1.0 + np.exp(-x))


def tanh(x: np.ndarray) -> np.ndarray:
    return np.tanh(x)


def relu(x: np.ndarray) -> np.ndarray:
    return np.maximum(0.0, x)


def leaky_relu(x: np.ndarray, alpha: float = 0.1) -> np.ndarray:
    return np.where(x > 0.0, x, alpha * x)


def softplus(x: np.ndarray) -> np.ndarray:
    return np.log1p(np.exp(x))


def main() -> None:
    x = np.linspace(-8, 8, 500)

    plt.figure(figsize=(10, 6))
    plt.plot(x, sigmoid(x), label="Sigmoid")
    plt.plot(x, tanh(x), label="Tanh")
    plt.plot(x, relu(x), label="ReLU")
    plt.plot(x, leaky_relu(x), label="Leaky ReLU (alpha=0.1)")
    plt.plot(x, softplus(x), label="Softplus")

    plt.title("Common Activation Functions")
    plt.xlabel("Input x")
    plt.ylabel("f(x)")
    plt.axhline(0, color="black", linewidth=0.6)
    plt.axvline(0, color="black", linewidth=0.6)
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.legend()
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    main()

```
