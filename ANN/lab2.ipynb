```python
class mcCulloch:
    def __init__(self, weights, threshold):
        self.weights = np.array(weights)
        self.threshold = threshold

    def andnot(self, x):
        s = np.dot(x, self.weights)

        # binary step activation function 
        if(s >= self.threshold) :
            return 1 
        else : 
            return 0
            
    def and_(self, x):
        s = np.dot(x, self.weights)

        # binary step activation function 
        if(s >= self.threshold) :
            return 1 
        else : 
            return 0
            
    def or_(self, x):
        s = np.dot(x, self.weights)

        # binary step activation function 
        if(s >= self.threshold) :
            return 1 
        else : 
            return 0

input = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

print ("""
ANN Lab 2  
Dipanshu Ambilkar (A-49)
=================================================
""")
neuron = mcCulloch(weights=[1, -1], threshold=1)

print("X1 X2 -> ANDNOT Output")
for x in inputs:
    print(x[0], " ",x[1],"    ", " ->    ", neuron.andnot(x))

print("=================================================")
    
neuron = mcCulloch(weights=[1, 1], threshold=2) 

print("X1 X2  -> AND Output")
for x in inputs:
    print(x[0], " ",x[1], " ->    ", neuron.and_(x))

print("=================================================")
    
neuron = mcCulloch(weights=[1, 1], threshold=1) 

print("X1 X2  -> OR Output")
for x in inputs:
    print(x[0], " ",x[1], " ->    ", neuron.or_(x))

```
