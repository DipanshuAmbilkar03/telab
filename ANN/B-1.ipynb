```python
"""B-1: Back Propagation Network for XOR (binary input and output)."""

import numpy as np


def sigmoid(x: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-x))


def sigmoid_grad(a: np.ndarray) -> np.ndarray:
    return a * (1 - a)


def main() -> None:
    X = np.array(
        [
            [0, 0],
            [0, 1],
            [1, 0],
            [1, 1],
        ],
        dtype=float,
    )
    y = np.array([[0], [1], [1], [0]], dtype=float)

    rng = np.random.default_rng(2)
    W1 = rng.normal(0, 1, size=(2, 2))
    b1 = np.zeros((1, 2))
    W2 = rng.normal(0, 1, size=(2, 1))
    b2 = np.zeros((1, 1))

    lr = 0.5
    epochs = 10000

    for epoch in range(1, epochs + 1):
        z1 = X @ W1 + b1
        a1 = sigmoid(z1)
        z2 = a1 @ W2 + b2
        y_hat = sigmoid(z2)

        loss = np.mean((y - y_hat) ** 2)

        # Backpropagation.
        d_out = (y_hat - y) * sigmoid_grad(y_hat)
        dW2 = a1.T @ d_out / len(X)
        db2 = np.sum(d_out, axis=0, keepdims=True) / len(X)

        d_hidden = (d_out @ W2.T) * sigmoid_grad(a1)
        dW1 = X.T @ d_hidden / len(X)
        db1 = np.sum(d_hidden, axis=0, keepdims=True) / len(X)

        W2 -= lr * dW2
        b2 -= lr * db2
        W1 -= lr * dW1
        b1 -= lr * db1

        if epoch % 1000 == 0:
            print(f"Epoch {epoch:5d}, MSE Loss: {loss:.6f}")

    final_pred = (y_hat >= 0.5).astype(int)
    acc = np.mean(final_pred == y) * 100

    print("\nXOR Predictions after training:")
    for x_row, p in zip(X.astype(int), final_pred.flatten()):
        print(f"Input {tuple(x_row)} -> Output {p}")

    print(f"\nTraining accuracy: {acc:.2f}%")


if __name__ == "__main__":
    main()

```
